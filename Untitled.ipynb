## Необхідні для проєкту залежності

```python
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras.layers import Dense, Dropout
from keras.models import Sequential
```

## Набір даних

База даних MNIST (Модифікована база даних Національного інституту стандартів і технологій[1]) — це велика база даних рукописних цифр, яка зазвичай використовується для навчання різних систем обробки зображень.[2][3] База даних також широко використовується для навчання та тестування в області машинного навчання.[4][5] Він був створений шляхом «повторного змішування» зразків з оригінальних наборів даних NIST.[6] Творці вважали, що, оскільки навчальний набір даних NIST був узятий від співробітників Американського бюро перепису населення, а тестовий набір даних був узятий від американських старшокласників, він не підходить для експериментів з машинним навчанням.[7] Крім того, чорно-білі зображення від NIST були нормалізовані для розміщення в обмежувальній рамці 28x28 пікселів і згладжені, що ввело рівні сірого.[7]

```python
from keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train.shape
```

```python
y_train.shape
```

## Візуалізація набору даних

```python
f, ax = plt.subplots(1, 10, figsize=(28, 28))

for i in range(10):
    sample = x_train[y_train == i][0]
    ax[i].imshow(sample, cmap='gray')
    ax[i].set_title(f'Label {i}')
```

```python
for i in range(10):
    print(i)
```

## Константи

```python
max_pixel_brightness = 255
```

## Нормалізація даних

```python
y_train_categorical = keras.utils.to_categorical(y_train, 10)
y_test_categorical = keras.utils.to_categorical(y_test, 10)
```

```python
# нормалізовані дані від 0 до 1 - покращує навчання нейронної мережі
x_train_normalized = x_train / max_pixel_brightness
x_test_normalized = x_test / max_pixel_brightness
```

```python
# реструктуризація потрібна для того щоб подати значення яскравості пікселів на вхід нейронної мережі = 764 вхідних нейрони
x_train_reshaped = x_train_normalized.reshape(x_train_normalized.shape[0], -1)
x_test_reshaped = x_test_normalized.reshape(x_test_normalized.shape[0], -1)
```

```python
x_train_reshaped.shape
```

```python
x_test_reshaped.shape
```

## Програмування моделі для розпізнавання чисел написаних від руки

```python
model = Sequential()

# по 128 нейронів у прихованих шарах нейронної мережі
# relu показує кращі результати тренуванн ніж sigmoid та tanh
model.add(Dense(units=128, input_shape=(784,), activation="relu"))
model.add(Dense(units=128, activation="relu"))
# для запобігання overfitting'у
model.add(Dropout(0.25))
# softmax для проблем класифікації - вихід від 0 до 1
model.add(Dense(units=10, activation="softmax"))

# categorical_crossentropy використовується для проблем багатокласової класифікації
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
model.summary()
```

## Тренування моделі із заздалегідь підготованим набором даних

```python
history = model.fit(x=x_train_reshaped, y=y_train_categorical, batch_size=512, epochs=10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['loss'])
plt.title('Точність моделі')
plt.ylabel('Точність')
plt.xlabel('Епоха')
plt.legend(['Точність', 'Втрати'])
plt.show()
```

## Застосування моделі до тестового набору даних

```python
test_loss, test_accuracy = model.evaluate(x_test_reshaped, y_test_categorical)
f"Втрати: {test_loss}, Точність: {test_accuracy}"
```

```python
y_pred = model.predict(x_test_reshaped)
y_pred_classes = np.argmax(y_pred, axis=1)
```

```python
random_idx = np.random.choice(len(x_test_reshaped))
x_sample = x_test_reshaped[random_idx]
y_true = np.argmax(y_test_categorical, axis=1)
y_sample_true = y_true[random_idx]
y_sample_pred_class = y_pred_classes[random_idx]

plt.title(f'Predicted: {y_sample_pred_class}, True: {y_sample_true}')
plt.imshow(x_sample.reshape(28, 28), cmap='gray')
```
